```python
import pandas as pd

df = pd.read_csv("Coffee_Sales/index_1.csv")

df.head()
```

## Quick Data EDA 

```python
df.shape
```

```python
# exploring and checking data for nulls
df.info()

# column "card" has few count since it will be blank since it was paid in cash
```

### Filling nulls in 'card' column:

```python
# If na then add "no card"
df['card'] = df['card'].fillna("No Card")

df.info()
```

### Extracting 'time' from 'datetime' in order to bin trameframeas as categories:

```python
# extract time from date time column
df['time'] = df['datetime'].str.split(" ", expand=True)[1]

df
```

```python
# Creating bins into type of hours 

bins   = [0, 5, 8, 11, 14, 17, 20, 24]   # hour edges or ranges
labels = ['Late Night','Early Morning','Morning',
          'Lunch/Midday','Afternoon','Evening',"Mid-Night"]

df['hour'] = pd.to_datetime(df['time'].astype(str)).dt.hour
df['time_category'] = pd.cut(df['hour'], bins=bins, labels=labels, right=False)

df
```

### Marking 'card type payments' as '1' in order to aggregate analysis better:

```python
df.cash_type.unique()
```

```python
# extracting transactions that were card payment to 1.
df['is_card'] = (df['cash_type'] == 'card').astype(int)

df
```

### Finding the operating hours of the vending machine:

```python
# what are the store working hours?

min = df['time'].min()
max = df['time'].max()

print('opening:', min, '\nclosing:', max)
```

### Converting 'dates' into 'day of the week' to aggregate analysis better:

```python
# naming day of the week using the date.
df['day_of_the_week'] = pd.to_datetime(df['date'])
df['day_of_the_week'] = df['day_of_the_week'].dt.day_name()

df
```

### Marking price ranges in bins (Cheap, Average, Expensive):

```python
# exploring price ranges of the cafe
min_price = df['money'].min()
max_price = df['money'].max()

print('min price:', min_price, '\nmax price:', max_price,'\n unique prices:', df['money'].unique())
```

```python
# Creating prices into bins price categories. 

money_bins = [0,26,35,100]
money_labels = ["Cheap", "Average", "Expensive"]

df['price_category'] = pd.cut(df['money'], bins=money_bins, labels=money_labels, right=False)
df
```
